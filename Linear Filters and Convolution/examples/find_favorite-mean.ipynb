```matlab
addpath('../');
```

```matlab
% load an image (shelf)
I = rgb2gray(imread('../pictures/shelf.png'));
imshow(I);
```

```matlab
% load the template to search for
 t = rgb2gray(imread('../pictures/template.jpg'));
 imshow(t);
```

```matlab
% calculate the cross-correlation
c = xcorr2(t, I);

% display the surface
surf(c), shading flat;
```

```matlab
% find indices of the place where the template matches
[x_index, y_index] = find_favorite(t, I)
```

```matlab
% draw a rectangle where the template is supposed to be located
imshow(I), imrect(gca, [x_index, y_index, size(t, 2), size(t, 1)]);
```

```matlab
% find the mean of the template
mean_template = mean2(t)
```

```matlab
% subtract it from the template, so it has zero mean
t_zero_mean = t - mean_template;
```

```matlab
% calculate the cross-correlation
c = xcorr2(t_zero_mean, I);

% display the surface
surf(c), shading flat;
```

```matlab
% subtract mean from the image
I_zero_mean = I - mean2(I);

% calculate the location of the template
[x_index, y_index] = find_favorite(t_zero_mean, I_zero_mean)

% draw a rectangle where the template is supposed to be located
imshow(I_zero_mean), imrect(gca, [x_index, y_index, size(t_zero_mean, 2), size(t_zero_mean, 1)]);
```

```matlab
% load a dark shelf image
I = rgb2gray(imread('../pictures/dark shelf.jpg'));
imshow(I);
```

```matlab
% calculate the cross-correlation (normalized)
c = normxcorr2(t, I);

% display the surface
surf(c), shading flat;
```

```matlab
% find the peak
[y_peak, x_peak] = find(c == max(c(:)))

% and get indices
x_index = x_peak - size(t, 2) + 1
y_index = y_peak - size(t, 1) + 1
```

```matlab
% draw a rectangle
imshow(I), imrect(gca, [x_index, y_index, size(t, 2), size(t, 1)]);
```
