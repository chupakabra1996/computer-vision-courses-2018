```matlab
addpath('../');
```

```matlab
cat = rgb2gray(imread('../pictures/cat.jpg'));
dog = rgb2gray(imread('../pictures/dog.jpg'));
figure, imshow(cat);
figure, imshow(dog);
```

```matlab
dog_low_freq = imboxfilt(dog, 23);
imshow(dog_low_freq);
```

```matlab
cat_high_freq = histeq(cat - imboxfilt(cat, 23));
imshow(cat_high_freq);
```

```matlab
imshow(cat_high_freq ./ 10 + dog_low_freq ./ 1.1);
```
