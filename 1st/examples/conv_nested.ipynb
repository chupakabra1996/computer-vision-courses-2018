```matlab
% get a real image
I = rgb2gray(imread('../pictures/cereals.png'));
imshow(I);
```

```matlab
% generate a gaussian 3x3 kernel
gauss_kernel = fspecial('gaussian', 3)
```

```matlab
% convolute an image with gaussian kernel
J = conv_nested(I, gauss_kernel);
imshow(J);
```
